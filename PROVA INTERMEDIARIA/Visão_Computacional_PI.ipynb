**PROBLEMA 2 (2.0 PONTOS).** Considere a função abaixo, que recebe quatro parâmetros: uma imagem em níveis de cinza, a coordenada de uma pixel (x,y) e um valor T.

```
def enigma(img,x,y,T):
  if abs(img[x+1][y]-img[x][y])>=T:
    return 255
  else:
    return 0
```

Utilize esta função para calcular um contorno da imagem lena.png, lida no código abaixo:

```
import cv2 as cv
from google.colab.patches import cv2_imshow
img = cv.imread('lena.png')
print(img.shape)
cv2_imshow(img)
```

```
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY).astype(np.uint8)
img_T = img_gray.copy()
rows, cols = img_T.shape
for i in range(1,rows-1):
  for j in range(1,cols-1):
    img_T[i][j] = enigma(img_gray, i, j, 200)
cv2_imshow(img_T)
```

Usando o frame e o background abaixo, implemente a operação de subtração de background em OpenCV.

```
import cv2 as cv
from google.colab.patches import cv2_imshow
frame = cv.imread('frame.png')
cv2_imshow(frame)
back = cv.imread('background.png')
cv2_imshow(back)
```

```
#implemente sua solução aqui
def sub_imgs(img1, img2, T):
  img_f = img1.copy()
  if img1.shape != img2.shape:
    print("Erro. Tamanho diferente.")
    return
  r, c = img1.shape
  for i in range(r):
    for j in range(c):
      if abs(img1[i][j] - img2[i][j]) > T:
        img_f[i][j] = 255
      else:
        img_f[i][j] = 0
  return img_f

frame_gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
back_gray = cv.cvtColor(back, cv.COLOR_BGR2GRAY)
cv2_imshow(np.hstack((frame_gray,back_gray)))
img_f = sub_imgs(frame_gray*0.4, back_gray*0.4, 15)
cv2_imshow(img_f)
```

**PROBLEMA 4 (3.0 PONTOS).** Abaixo, temos um código em OpenCV que abre uma imagem BGR com três canais de cor, cada cor representada como um número inteiro. 

O sistema de cores CMYK é um sistema de cores subtrativo formado pelas cores Ciano (C), Magenta (M), Amarelo (Y) e Preto (K) e, estas cores, trabalham no intervalo [0.0,1.0]. Abaixo, temos as fórmulas de conversão de RGB para CMYK:


*   K=1-max(R,G,B)/255
*   C=(1-K-R/255)/(1-K)
*   M=(1-K-G/255)/(1-K)
*   Y=(1-K-B/255)/(1-K)

Escreva um programa em OpenCV que converta a imagem BGR para CMYK. Imprima o shape da imagem convertida e exiba os canais gerados.

```
import cv2 as cv
from google.colab.patches import cv2_imshow
img = cv.imread('lena.png')
print(img.shape)
cv2_imshow(img)
```

```
#implemente sua solução aqui
def rgb_to_cmyk(img):
  img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  rows, cols, channels = img_rgb.shape
  img_cmyk = np.zeros((rows,cols, 4))
  for i in range(rows):
    for j in range(cols):
      R, G, B = img_rgb[i,j, :]
      K = 1 - max((R, G, B))/255
      C = (1 - K - R/255)/ (1 - K)
      M = (1 - K - G/255)/ (1 - K)
      Y = (1 - K - B/255)/ (1 - K)
      img_cmyk[i,j] = np.array([C,M,Y,K])
  return img_cmyk

      

img_cmyk = rgb_to_cmyk(img)
```

```
print(img_cmyk.shape)
```

```
C, M, Y, K = cv.split(img_cmyk)
C, M, Y, K = C*255, M*255, Y*255, K*255
cv2_imshow(np.hstack( (C, M, Y, K) ))
```
